**1.1 머신러닝이란?**

머신러닝 = 명시적인 프로그래밍 없이 컴퓨터가 학습능력을 갖게 연구하는 분야(PTE의 반복)

**1.2 왜 머신러닝을 사용하나요?**

사용 이유 :
1. 기존 = 문제가 어렵고, 기존 규칙을 알아내고 작성을 반복하는 과정은 비효율적, 규칙이 바뀔 때마다 대응 어려움 /  머신러닝 = 훈련시키면 유지 보수가 쉽고 정확도가 높아짐, 새로운 규칙 빠르게 인식 가능


2. 너무 복잡하거나 알고리즘이 없는 문제 -> 예상치 못한 상관관계로 새로운 추세 발견 가능(문제 더 잘 이해하도록 도움)

데이터 마이닝 : 대용량 데이터를 분석하여 숨겨진 패턴을 발견하는 것

**<머신러닝 분류>**

**지도학습**

일반적으로 분류문제에 활용(ex.로지스틱 회귀사용)

-수학적 표현
**비지도학습**

훈련데이터에 레이블이 없는 경우
-군집알고리즘을 이용
-비슷한 자료끼리 묶어 class만들어 분류

시각화 알고리즘 이용

특성추출을 이용한 차원 축소

이상치 탐지

연광 규칙 학습

**준지도학습**

레이블이 일부만 있는 데이터에 활용
-레이블이 없는 데이터에 데이터에 레이블 부여
-레이블이 부여된 데이터를 이용해 학습

**자기지도학습**

레이블이 전혀없는 데이터에서 활용

전체 데이터셋에 레이블 부여 후 학습

**강화학습**

에이전트가 환경을 인식하고 행동을 하는데 보상도는 벌점을 통해 최상의 전략을 스스 학습하는 방식

로봇에 많이 활용됨.

image.png

**배치학습 / 온라인 학습**-점진적으로 학습할 수 있는지 여부

**배치학습** : 시간과 자원 많이 소모 - 오프라인 학습 진행 - 성능이 시간에 따라 감소함.(=모델부패,데이터 드리프트)
많은 단점이 존재 보완하기 위해선 많은 비용이 듬.

**온라인 학습** : 데이터를 수차적으로 또는 묶음단위로 시스템 훈련시킴. 데이터 도착즉시 즉가적으로 빠르게 학습 가능.(주식시장과 같이 새로운 패턴을 감지할 때 유용)
ex 외부메모리학습, 이상치 탐지 알고리즘

-이때 학습률이 중요

단점 : 학습률이 높으면 기억 오래 하지 못함, 학습률이 낮으면 느리게 학습되고 잡음이나 데이터 포인트에 덜 민감해짐, **시스템에 나쁜 데이터가 주입되었을 때 성능 감소함**- 즉각 학습 중지 시켜야함.

**사례기반학습 / 모델기반학습**-어떻게 일반화 되는지에 따라 분류

**사례기반학습** : 훈련 샘플 기억해 학습(-유사도측정을 통해 새로운 데이터와 학습한 샘플 비교)

**모델기반학습** : 샘플들의 모델을 만들어 예측에 사용하는것 - 선형 회귀 알고리즘 등장(훈련데이터 공급- 데이터에 가장 잘 맞는 선형 모델의 파라미터를 찾음)



```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn. linear_model import LinearRegression
# 데이터를 다운로드
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]]. values
y = lifesat[["Life satisfaction"]]. values
# 데이터를 그래프로 나타냄
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
# 선형 모델을 선택
model = LinearRegression()
# 모델을 훈련
model. fit (X, y)
# 키프로스에 대해 예측을 만듦
X_new = [[37_655.2]]# 2020년 키프로스 1인당 GDp
print(model.predict(X_new)) # 출력: [[6.30165767]]
```

선형 모델을 k-최근접 이웃 회귀로 바꾸면

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
# 데이터를 다운로드
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]]. values
y = lifesat[["Life satisfaction"]]. values
# 데이터를 그래프로 나타냄
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()
# 선형 모델을 선택
model = KNeighborsRegressor(n_neighbors=3)
# 모델을 훈련
model. fit (X, y)
# 키프로스에 대해 예측을 만듦
X_new = [[37_655.2]]# 2020년 키프로스 1인당 GDp
print(model.predict(X_new)) # 출력: [[6.30165767]]
```

#머신러닝 주요 도전 과제

**나쁜 데이터 사례**
1. 충분하지 않은 양의 훈련 데이터(ex믿기힘든 데이트 효과)
2. 대표성 없는 훈련 데이터 - 샘플링 편향(ex루스벨트대통령선거)
3. 낮은 품질 데이터- 오류,이상치, 잡음
4. 관련성 없는 특성

**나쁜 알고리즘 사례**
1. 훈련데이터 과대적합(= 과도한 일반화)

해결법 - 파라미터수가 적은 모델 선택, 훈련데이터 특성수 줄이기, 모델제약 가해 단순화(=**제약**), 훈련데이터 더 모음, 훈련데이터 잡음 줄임(오류데이터 수정과 이상치 제거), 자유도를 학습 알고리즘에 부여, 하이퍼파라미터 튜닝

하이퍼파라미터 : 학습하는 동안 적용할 규제의 양 결정(=학습 알고리즘의 파라미터)

2. 훈련 데이터 과소적합 = 모델이 너무 단순해 데이터 내제된 구조를 학습하지 못함

해결법 - 모델 파라미터가 더 많은 강력한 모델 선택, 학습 알고리즘에 더 좋은 특성을 제공, 모델 제약 줄이기(규제 하이퍼파라미터 감소)


#테스트와 검증

샘플이 잘 일반호 되었는지 확인 방법 = only 새로운 샘플에 실제로 적용
-> 훈련데이터를 훈련세트와 테스트 세트로 구분하는 것이 좋음

일반화 오차,외부 샘플 오차 : 새로운 샘플에 대한 오차 비율

추정값이 모델이 얼마나 잘 작동하는지 알려줌.

.

**하이퍼파라미터 튜닝과 모델 선택**

모델 평가 = just 테스트 세트 사용하면 됨

but 세트모델에 최적화 된 모델을 만들어 새로운데이터에 잘 작동하지 않을 수 있음 이때 사용하는게 홀드아웃 검정.

**홀드아웃 검정** : 훈련데이터에 일부를 때어 여러 후보 모델을 평가하고 가장 좋은 모델 선택(=검증세트,개발세트, 데브세트)

1. 훈련세트에서 다양한 하이퍼파라미터 값을 가진 여러 모델 훈련
2. 검증 세트에서 가장 높은 성능을 내는 모델 선택
3. 홀드아웃 검증 과정이 끝나면 최선의 모델을 전체 훈련 세트에서 다시 훈련 시켜 최종 모델을 만듬
4. 최종모델을 테스트 세트에서 평가하여 일반화 오차를 추정

단점 : 검증 세트가 너무 적으면 정확하게 평가 되지 않고 너무 크면 남은 훈련데이터가 전체 훈련데이트 세트 보다 훨씬 작아짐 - **교차검증을 통해 해결**

.


**데이터 불일치**

검증 세트와 데이터 세트는 대표성을 가진 사징으로만 구성해야함.

**훈련-개발세트**에서 모델 훈련시키고 평가 - 잘 작동하지 않으면 과대적합

공짜점심없음이론 : 데이터에 관해 완벽하게 아무런 가정을 하지 않으면 한 모델을 다른 모델보다 선호할 근거 없음
