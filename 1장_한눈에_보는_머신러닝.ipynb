**<머신러닝 정의>**

머신러닝 = 명시적인 프로그래밍 없이 컴퓨터가 학습능력을 갖게 연구하는 분야(PTE의 반복)

사용 이유 : 기존 규칙을 알아내고 작성을 반복하는 과정은 비효율적이지만 머신러닝을 훈련시키면 유지 보수가 귑고 정확도가 높아짐

데이터 마이닝 : 대용량 데이터를 분석하여 숨겨진 패턴을 발견하는 것

**<머신러닝 분류>**

**비지도학습**

훈련데이터에 레이블이 없는 경우
-군집알고리즘을 이용
-비슷한 자료끼리 묶어 class만들어 분류

**준지도학습**

레이블이 일부만 있는 데이터에 활용
-레이블이 없는 데이터에 데이터에 레이블 부여
-레이블이 부여된 데이터를 이용해 학습

**자기지도학습**

레이블이 전혀없는 데이터에서 활용

전체 데이터셋에 레이블 부여 후 학습
